```python
import matplotlib.pyplot as plt

# Model performance data
v1_metrics = {
    "test_mse": 0.062023283810268046,
    "test_mae": 0.19785151442570448,
    "test_r2": 0.28799993904110577,
    "val_mse": 0.05912314158899554,
    "val_mae": 0.1859429632849997,
    "val_r2": 0.21520800569017207
}

v2_metrics = {
    "test_mse": 0.05745355406154862,
    "test_mae": 0.1853406882206114,
    "test_r2": 0.340458429784791,
    "val_mse": 0.05659781950060293,
    "val_mae": 0.17838794699350374,
    "val_r2": 0.24872876430955448
}

# Prepare data for plotting
metrics = ['MSE', 'MAE', 'R^2']
v1_test = [v1_metrics['test_mse'], v1_metrics['test_mae'], v1_metrics['test_r2']]
v1_val = [v1_metrics['val_mse'], v1_metrics['val_mae'], v1_metrics['val_r2']]
v2_test = [v2_metrics['test_mse'], v2_metrics['test_mae'], v2_metrics['test_r2']]
v2_val = [v2_metrics['val_mse'], v2_metrics['val_mae'], v2_metrics['val_r2']]

# Plotting
x = range(len(metrics))  # the label locations
width = 0.2  # the width of the bars

colors = ['#fdae6b', '#fd8d3c', '#6baed6', '#3182bd',  '#74c476', '#31a354'] # orange x2, blue x2, green x2
fig, ax = plt.subplots()
rects1 = ax.bar(x, v1_test, width, label='v1 Linear Regression - Test', color=colors[0])
rects2 = ax.bar([p + width for p in x], v1_val, width, label='v1 Linear Regression - Validation', color=colors[1])
rects3 = ax.bar([p + width*2 for p in x], v2_test, width, label='v2 Random Forest - Test', color=colors[2])
rects4 = ax.bar([p + width*3 for p in x], v2_val, width, label='v2 Random Forest - Validation', color=colors[3])

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Model Performance by Metric and Dataset')
ax.set_xticks([p + width for p in x])
ax.set_xticklabels(metrics)
ax.legend()

# Function to add labels on bars
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(round(height, 3)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center',
                    va='bottom',
                    fontsize=8
        )

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

fig.tight_layout()

plt.show()
```
